```python

# 1. Data Gathering
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
train = pd.read_csv("tested.csv")
print(train.head())

# 2. Data Preprocessing
print(train.describe())
print(train.isnull().sum())
train_x = train[["Pclass", "Sex"]]
train_y = train[["Survived"]]
train_x["Sex"].replace("male", 1, inplace=True)
train_x["Sex"].replace("female", 0, inplace=True)
print(train_x.head())
tr_x, cv_x, tr_y, cv_y = train_test_split(train_x, train_y, test_size=0.30)

print(tr_x.head())
print(tr_y.head())

# 3. Choosing and Training a Model
rf = RandomForestClassifier()
rf.fit(tr_x, tr_y)

# 4. Evaluating the Model
Accuracy_RandomForest = rf.score(cv_x, cv_y)
print("Accuracy of random forest = {:.2f}%".format(Accuracy_RandomForest * 100))

lgr = LogisticRegression()
lgr.fit(tr_x, tr_y)
Accuracy_LogisticRegression = lgr.score(cv_x, cv_y)
print("Accuracy of logistic regression = {:.2f}%".format(Accuracy_LogisticRegression * 100))

# 6. Predict the Data
test = pd.read_csv("test.csv")
test.head()
test_x = test[["Pclass", "Sex"]]
test_x.head()
test_x["Sex"].replace("male", 1, inplace=True)
test_x["Sex"].replace("female", 0, inplace=True)
test_x.head()
prd = rf.predict(test_x)
prd
op = test[['PassengerId']]
op['Survived'] = prd
op.head()
op.to_csv("Submission.csv", index=False)
op.tail()

```
